### Lab Task 1:
1.	Implement the step function as an activation function in Python using both the math library and TensorFlow. Test it with different input values and visualize its output using a simple plot.

#### USING MATH LIBRARY

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def step_math(s):
    return 1 if s >= 0 else 0

inputs = np.arange(-10, 11)
outputs_math = [step_math(s) for s in inputs]

plt.plot(inputs, outputs_math)
plt.grid()
plt.show()
```

#### USING TENSORFLOW LIBRARY 

```python
import tensorflow.compat.v1 as tf
import numpy as np
import matplotlib.pyplot as plt

def step_tf(x):
    return tf.cast(tf.greater_equal(x, 0), tf.int32)

x = tf.placeholder(tf.float32, name='x')
step_output = step_tf(x)

with tf.Session() as sess:
    inputs_tf = np.arange(-10, 11)
    outputs_tf = sess.run(step_output, feed_dict={x: inputs_tf})

plt.plot(inputs_tf, outputs_tf)
plt.grid(True)
plt.show()
```

### Lab Task 2:
2.	Create a Python function for the ReLU (Rectified Linear Unit) activation function using both the math library and TensorFlow. Test it with a range of input values and plot its output.

#### USING MATH LIBRARY 

```python
import math
from matplotlib import pyplot
def ReLU(s):
    return (max(0,s))

inputs = [s for s in range(-10,10)]
outputs = [ReLU(s) for s in inputs]
pyplot.plot(inputs,outputs)
pyplot.grid()
pyplot.show()
```

#### USING TENSORFLOW LIBRARY

```python
import tensorflow.compat.v1 as tf

x = tf.placeholder(tf.float32, name = 'x')
ReLU_output = ReLU(x)

with tf.Session() as sess:
    input_value = inputs = [x for x in range(-10,10)]
    result = sess.run(ReLU_output, feed_dict = {x: input_value})
    print("ReLU of\n", input_value, " \nis\n", result)

pyplot.plot(input_value, result)
pyplot.grid()
pyplot.show()
```

### Lab Task 3:
3.	Write a Python code to implement the Sigmoid activation function using both the math library and TensorFlow. Visualize its output and discuss its characteristics compared to other activation functions.

#### USING MATH LIBRARY

```python
import math
from matplotlib import pyplot
def sigmoid(s):
    return 1/(1 + math.exp(-s))

inputs = [s for s in range(-10,10)]
outputs = [sigmoid(s) for s in inputs]
pyplot.plot(inputs,outputs)
pyplot.grid()
pyplot.show()
```

#### USING TENSORFLOW LIBRARY

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

x = tf.placeholder(tf.float32, name = 'x')

sigmoid_output = tf.sigmoid(x)

with tf.Session() as sess:
    input_value = inputs =[x for x in range (-10,10)]
    result = sess.run(sigmoid_output, feed_dict={x:input_value})
    print("Sigmoid of \n",input_value,"\nis\n",result)

pyplot.plot(input_value, result)
pyplot.grid()
pyplot.show()
```

### Lab Task 4:
4.	Develop a Python function for the Tanh (Hyperbolic Tangent) activation function using both the math library and TensorFlow. Test it with various input values and plot its output to observe its behavior.

#### USING MATH LIBRARY

```python
import math
from matplotlib import pyplot
def TanH(s):
    return (math.exp(s)-math.exp(-s))/(math.exp(s) + math.exp(-s))

inputs = [s for s in range(-10,10)]
outputs = [TanH(s) for s in inputs]
pyplot.plot(inputs,outputs)
pyplot.grid()
pyplot.show()
```

#### USING TENSORFLOW LIBRARY

```python
import tensorflow.compat.v1 as tf

def tanH(x):
    return (tf.exp(x) - tf.exp(-x)) / (tf.exp(x) + tf.exp(-x))
x = tf.placeholder(tf.float32, name = 'x')
tanH_output = tanH(x)

with tf.Session() as sess:
    input_value = inputs = [x for x in range(-10,10)]
    result = sess.run(tanH_output, feed_dict = {x: input_value})
    print("tanH of\n", input_value, " \nis\n", result)

pyplot.plot(input_value, result)
pyplot.grid()
pyplot.show()
```

### Lab Task 5:
5.	Implement the Leaky ReLU activation function in Python using both the math library and TensorFlow. Experiment with different leakage coefficients and plot the function's output.

#### USING MATH LIBRARY

```python
import math
from matplotlib import pyplot
def LeakyReLU(s,a=0.1):
    return (max(a*s,s))

inputs = [s for s in range(-10,10)]
outputs = [LeakyReLU(s) for s in inputs]
pyplot.plot(inputs,outputs)
pyplot.grid()
pyplot.show()
```

#### USING TENSORFLOW LIBRARY

```python
import tensorflow.compat.v1 as tf

def LeakyReLU(x):
    alpha = 0.1
    return tf.maximum(alpha * x, x)

x = tf.placeholder(tf.float32, name = 'x')
LeakyReLU_output = LeakyReLU(x)

with tf.Session() as sess:
    input_value = inputs = [x for x in range(-10,10)]
    result = sess.run(LeakyReLU_output, feed_dict = {x: input_value})
    print("LeakyReLU of\n", input_value, " \nis\n", result)

pyplot.plot(input_value, result)
pyplot.grid()
pyplot.show()
```

### Lab Task 6:
6.	Create a Python function for the Softmax activation function using both the math library and TensorFlow.

#### USING MATH LIBRARY

```python
import math
from matplotlib import pyplot

def softmax(inputs):
    exp_values = [math.exp(x) for x in inputs]
    sum_exp_values = sum(exp_values)
    softmax_values = [exp_val / sum_exp_values for exp_val in exp_values]
    return softmax_values

inputs = [s for s in range(-10, 10)]
outputs = softmax(inputs)

pyplot.plot(inputs, outputs)
pyplot.grid()
pyplot.show()
```

#### USING TENSORFLOW LIBRARY

```python
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
tf.compat.v1.disable_eager_execution()
def plot_softmax(inputs):
    softmax_values = tf.nn.softmax(inputs)
    return softmax_values

inputs = np.array([s for s in range(-10, 10)], dtype=np.float32)
softmax_values = plot_softmax(inputs)

with tf.compat.v1.Session() as sess:
    outputs = sess.run(softmax_values)
    print("Softmax value of\n", inputs, " \nis\n", outputs)

plt.plot(inputs, outputs)
plt.grid()
plt.show()
```
